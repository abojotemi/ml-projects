```python
import numpy as np
import pandas as pd
from sklearn.calibration import LabelEncoder
```

```python
housing = pd.read_csv('Housing.csv')
```

```python
X = housing.iloc[:,1:]
y = housing.iloc[:,0]
```

```python
X
```

```python
xe = LabelEncoder()
for i in range(5,10):
    X.iloc[:,i] = xe.fit_transform(X.iloc[:,i])
for i in range(1,3):
    X.iloc[:,-i] = xe.fit_transform(X.iloc[:,-i])

X.mainroad = xe.fit_transform(X.mainroad)
```

```python
X
```

```python
y = y.to_numpy().reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=np.random.randint(0, 100000))
```

```python
from sklearn.linear_model import LinearRegression


reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred = (reg.predict(X_test).reshape(-1,1) // 1)
```

```python
[y_pred,y_test]
```

```python
np.hstack((y_pred,y_test))
```

```python

from sklearn.metrics import mean_absolute_error


mean_absolute_error(y_test, y_pred)
```
